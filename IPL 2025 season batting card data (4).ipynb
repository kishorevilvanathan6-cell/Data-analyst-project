```python
data=df.groupby("Team")["Runs"].sum().sort_values()
data
```

```python
data.plot(kind="bar")
plt.title("MOST RUN SCORED BY TEAM IN IPL 2025")
plt.xlabel("TEAM")
plt.ylabel("RUN SCORED")
```

```python
data2=df.groupby("Team")["Runs"].sum()
data2
```

```python
data2.plot(kind="pie",autopct="%1.1f%%")
plt.title("More total Run scored in IPL 2025")
```
